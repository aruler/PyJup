# Let's dig into health expanses !

First, lets import what we need to import and dataset on which we will work futher on.
What we need is: 
1. seaborn  - for dataset and graph creation 
2. matplotlib.pyplot - for actually showing our graphs

We are using python to satisfy our curiosity hear in what you can see is jupyter notebook which you will probably open as me in Jupyter Lab

```python
import seaborn as sns
import matplotlib.pyplot as plt

healthexp = sns.load_dataset("healthexp")

print(healthexp.head(10))
```

What id like to see on this dataset at first is how average cost on health changed throught years

```python
mean_expenses_by_year = healthexp.groupby("Year")["Spending_USD"].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_expenses_by_year, x="Year", y="Spending_USD")
plt.title("mean expenses by year")
plt.xlabel("Year")
plt.ylabel("Mean Expenses")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

As we can see, there is almost linear growth on money spend on health.

Have you thought what would happen if we didn't collect data into mean and present it as it is in data set? Will it break cause we will have to many Y points for X point? Lets check that!

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=healthexp, x="Year", y="Spending_USD")
plt.title("Expenses by year")
plt.xlabel("Year")
plt.ylabel("Expenses")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Don't know about you, but i prefer the second one. Cause it also shows min-max diffrencces and how throught the years the gap between the least and the most money spending country goes bigger and bigger. But im also courious and by this point maybe you are too, which country is the max and which is the low, let's check that :)

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=healthexp, x="Year", y="Spending_USD",hue="Country")
plt.title("Country expenses by year")
plt.xlabel("Year")
plt.ylabel("Expenses")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Curiosity takes me further, let's see about life expectancy

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=healthexp, x="Year", y="Life_Expectancy",hue="Country")
plt.title("Country Life Expectancy by year")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
